```python
import numpy as np
import matplotlib.pyplot as plt


task1 =   True       ##Вывод самих КЛФ аналитически и кусочно
task2_1 = True     ##Вывод sin(x) до и после нелинейности
task2_2 = True     ##Вывод A*sin(x) после нелинейности (A = 1, 2, 2.5, 3)
task3_1 = True     ##Вывод sin(x), sin(x) после нелинейности и sin(x) после нелинейности и обратной
task3_2 = True     ##Вывод КЛФ + обратной с разными синусами
task3_3 = True     ##Вывод КЛФ + обратной с разными синусами
task4 =   True       ##Вывод самой двузначной КЛФ аналитически и кусочно
task5_1 = True     ##Вывод sin(x) до и после нелинейности
task5_2 = True     ##Вывод 2sin(x) до и после нелинейности
task5_3 = True     ##Вывод sin(x/B) до и после нелинейности, B = 2 (меняется строкой ниже)
B = 1/3

# Аналитическая функция
def f_analytical(x):
    return x + 0.5 * (-np.abs(x + 2) - np.abs(x + 1) + 
                      np.abs(x - 1) + np.abs(x - 2))


epslion = 10**-10

# Обратная функция
def y_obr(x):
    if (np.abs(x) < 1 - epslion):
        return -x
    return x


# Кусочно-линейная функция
def f_klf(x):
    y = np.piecewise(
        x,
        [x < -2,
         (-2 <= x) & (x < -1),
         (-1 <= x) & (x < 1),
         (1 <= x) & (x < 2),
         x >= 2],
        [lambda x: 1 * x + 3,
         lambda x: 0 * x + 1,
         lambda x: -1 * x + 0,
         lambda x: 0 * x - 1,
         lambda x: 1 * x - 3]
    )
    return y

# Дополнительные функции
def f_1(x):
    return 0.5 * np.sin(x)

def f_2(x):
    return 1.5 * np.sin(x)

def f_3(x):
    return 2.5 * np.sin(x)


def f_4(x):
    return 3 * np.sin(x)

# Создаём диапазон x
delta = 3.05
x = np.linspace(-delta, delta, 1000)

y_obrat = []
for val in x:
    y_obrat.append(y_obr(val))
plt.figure(figsize=(10, 6))
plt.plot(x, y_obrat, label='Аналитическая функция', linewidth=2)
plt.title('Обратная функция к однозначной КЛФ')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.legend()
plt.grid(True)
plt.show()


# Вычисляем значения функций
y_analytical = f_analytical(x)
y_klf = f_klf(x)
y_f1 = f_analytical(f_1(x))
y_f2 = f_analytical(f_2(x))
y_f3 = f_analytical(f_3(x))
y_f4 = f_analytical(f_4(x))
knots = [-2, -1, 1, 2]


y_f1_obr = []
for val in y_f1:
    y_f1_obr.append(y_obr(val))

y_f2_obr = []
for val in y_f2:
    y_f2_obr.append(y_obr(val))
    
y_f3_obr = []
for val in y_f3:
    y_f3_obr.append(y_obr(val))
    
y_f4_obr = []
for val in y_f4:
    y_f4_obr.append(y_obr(val))

```

```python
def f21_klf(x):
    y = np.piecewise(
        x, [x < -1, (-1 <= x) & (x < 1), (1 <= x) & (x < 2),
         (2 <= x) & (x < 4), x >= 4],
        [lambda x: 0 * x - 3, lambda x: 1 * x - 2, lambda x: 2 * x - 3,
         lambda x: 1 * x - 1, lambda x: 0 * x + 3]
    )
    return y

def f22_klf(x):
    y = np.piecewise(
        x,[x < -4, (-4 <= x) & (x < -2), (-2 <= x) & (x < -1),
         (-1 <= x) & (x < 1), x >= 1],
        [lambda x: 0 * x - 3, lambda x: 1 * x + 1, lambda x: 2 * x + 3,
         lambda x: 1 * x + 2, lambda x: 0 * x + 3]
    )
    return y


def f21_analytical(x):
    return 0.5 * (np.abs(x + 1) + np.abs(x - 1) - np.abs(x - 2) - np.abs(x - 4))

def f22_analytical(x):
    return 0.5 * (np.abs(x + 4) + np.abs(x + 2) - np.abs(x + 1) - np.abs(x - 1))


def f2_analytical(x, y):
    if y < 0:
        return f21_analytical(x)
    else:
        return f22_analytical(x)



delta2 = 4.5
x2 = np.linspace(-delta2, delta2, 1000)

y_klf21 = f21_klf(x2)
y_klf22 = f22_klf(x2)
y_analytical21 = f21_analytical(x2)
y_analytical22 = f22_analytical(x2)
```

```python
# Дополнительные функции для двузначной КЛФ
def f_21(x):
    return 1 * np.sin(x)

def f_22(x):
    return 2 * np.sin(x)

def f_21der(x):
    return 1 * np.cos(x)

def f_22der(x):
    return 2 * np.cos(x)

def f_23(x):
    return 1 * np.sin(x/B)

def f_23der(x):
    return 1/B * np.cos(x/B)



y_21 = f_21(x2)
y_21der = f_21der(x2)

y_22 = f_22(x2)
y_22der = f_22der(x2)

y_23 = f_23(x2)
y_23der = f_23der(x2)

y_f21 = []
y_f22 = []
y_f23 = []

for x, y in zip(y_21, y_21der):
    y_f21.append(f2_analytical(x,y))
    
for x, y in zip(y_22, y_22der):
    y_f22.append(f2_analytical(x,y))    
    
for x, y in zip(y_23, y_23der):
    y_f23.append(f2_analytical(x,y)) 


```
